```python
def distance(x1,x2):
    s = 0
    for i in range(len(x1)):
        s += (x1[i] - x2[i])**2
    return np.sqrt(s)
```

```python
def distance(x1,x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
def kni(x_train, xi, k):
    distances = []
    for i in range(len(x_train)):
        d = distance(xi,x_train[i])
        distances.append([d,i])
    distances.sort()
    distances = np.array(distances [:k])
    return distances[:,-1]
```

```python
l = [3,6,4,8,1,9,3,3,6,4,3]
freq = {}
for i in l:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
print(freq)
```

```python
def predict(x_train, y_train, X, k):
    pred = []
    for xi in X:
        knn_index = kni(x_train, xi, k)

        freq = {}

        for i in knn_index:
            i = int(i)
            if y_train[i] in freq:
                freq[y_train[i]] += 1
            else:
                freq[y_train[i]] = 1

        f = list(freq.keys())[0]
        for i in freq:
            if freq[i] > freq[f] :
                f = i
        pred.append(f)
        
    return pred
```

```python
y_train = y_train.reshape(-1)
```
